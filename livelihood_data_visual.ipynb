```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('FEH_00100107_200211034401.csv', encoding="cp932")
```

```python
data.columns
```

```python
data_re = data.drop(data.index[[0,1,2,3,4,54,53,52,51,50]])
```

```python
big = data_re[data_re["都市規模"]=="大都市"]
```

```python
big
```

```python
plt.rcParams["font.family"] = "AppleGothic"
name = big["暮らし向き"]
plt.rcParams['font.size'] = 16.0
plt.pie(big["value"],labels=name,counterclock=True,autopct="%1.1f%%")
```
